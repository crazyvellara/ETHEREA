```python
# outliner checking and removing df.boxplot?
# outliner= different data from our data range
# to find ouliners by df.boxplot
df.boxplot()
```

```python
# removing ooutliner(outlayers); only remove x by iqr(intel portal range)
#q3-q1 equation for iqr
#create a fn to remove
#q1-1.5*iqr-lower bond
#q3+1.5*iqr-upper bond
#
def remove_outliers_iqr(df, exclude_column):
  for col in df.columns:
    #skip the specific column
    if col == exclude_column:
      continue
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound =   Q1 - 1.5 * IQR
    upper_bound =   Q1 + 1.5 * IQR
    # filter the dataframe to remove rows with outliers in each column
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
  return df
  # specify the column to exclude
df_cleaned = remove_outliers_iqr(df, exclude_column='survived')



```

```python
# .quantile= to get the % in Q1 & Q2
df = df_cleaned
```

```python
# data scaling
# to get better model
from sklearn.preprocessing import MinMaxScaler
# select the features that you want to scale
features = ['pclass', 'fare', 'embark_town', 'who', 'adult_male', 'alone']
#create a MinMaxScaler object
scaler = MinMaxScaler()
#fit the scaler to your data and transform it
scaled_data = scaler.fit_transform(df[features])
# create a data form scaled_data with the correct column names
scaled_df = pd.DataFrame(scaled_data, columns=features)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = scaled_df[['pclass', 'fare', 'embark_town', 'who', 'adult_male', 'alone']]
y = df['survived'] # target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
# Train a basline logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

#predict and evaluate the model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Baseline model accuracy: {accuracy:.2f}')
```
